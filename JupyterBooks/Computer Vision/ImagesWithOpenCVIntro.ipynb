```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('mypic.jpg')
```

```python
cv.imshow("Display window", img)
cv.waitKey(0)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
cv.imshow("Gray Image", gray)
cv.waitKey(0)
cv. destroyAllWindows()
```

```python
noisereduced_version = cv.medianBlur(img,3)
cv.imshow("Noise Reduced", noisereduced_version)
cv.waitKey(0)
cv. destroyAllWindows()
```

```python
histogram_image = cv.calcHist(img, [0], None, [256],[0,256])
plt.hist(histogram_image.ravel(), 256, [0,256])
plt.show()
```

```python
color = ['b','g','r']
```

```python
for i, col in enumerate(color):
    hist = cv.calcHist(img, [i], None, [256], [0,256])
    plt.plot(hist, color = col)
    plt.xlim([0,256])
plt.show ()
```

```python
hist = cv.calcHist(img, [2], None, [256], [0,256])
plt.plot(hist, color = 'b')
plt.xlim([0,256])
```
