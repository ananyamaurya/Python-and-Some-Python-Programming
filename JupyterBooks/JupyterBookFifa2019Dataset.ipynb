```python
data_frame1 = pd.DataFrame(data_frame,columns=['Name','Wage','Value'])
data_frame1
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
wage = data_frame1['Wage'].replace('[\â‚¬,]','',regex = True).apply(value_to_float)
value = data_frame1['Value'].replace('[\â‚¬,]','',regex = True).apply(value_to_float)
data_frame1['Wage'] = wage
data_frame1['Value'] = value
data_frame1['Difference'] = data_frame1['Value'] - data_frame1['Wage']
data_frame1.sort_values('Difference',ascending=False)
```

```python
import seaborn as sns
sns.set()


graph = sns.scatterplot(x='Wage',y='Value',data=data_frame1)
graph
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool



TOOLTIPS = HoverTool(tooltips=[
    ("index","$index"),
    ("(Wage,Value)","(@Wage,@Value)"),
    ("Name","@Name")]
)


gra = figure(title='FIFA 2019',x_axis_label='Wage',y_axis_label='Value',plot_width=640, plot_height=640,tools=[TOOLTIPS])
gra.circle('Wage','Value',size=10,source= data_frame1)
show(gra)
```
